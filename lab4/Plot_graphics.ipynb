```python
import matplotlib.pyplot as plt
import numpy as np

f = open("load.log", 'r')
threads = list()
times = list()
queries = list()

for x in f:
    s = x.split(' ')
    queries.append(int(s[3]))
    times.append(float(s[4].split('\n')[0]))
    threads.append(int(s[0]))
```

```python
threads = np.array(threads)
times = np.array(times)
queries = list(map(str, queries))
```

```python
plot_x, plot_y = list(), list()
i_start = 0
i_end = 6 # max_num_of_queries
max_num_of_threads = 17
for i in range(1, max_num_of_threads):
    i_end = i_start+6*i
    plot_x.append(queries[i_start:i_end])
    plot_y.append(times[i_start:i_end])
    i_start = i_end
```

```python
plot_x_opt, plot_y_opt = list(), list()
i_start = len(queries) // 2
i_end = i_start + 6
for i in range(1, max_num_of_threads):
    i_end = i_start+6*i
    plot_x_opt.append(queries[i_start:i_end])
    plot_y_opt.append(times[i_start:i_end])
    i_start = i_end
```

```python
fig = plt.figure(figsize = (25, 20))
for x in range(16):
    ax = fig.add_subplot(4, 4, x+1)
    label1 = str(x+1) + " thread(s) before"
    label2 = str(x+1) + " thread(s) after"
    ax.bar(plot_x[x], plot_y[x], label = label1)
    ax.bar(plot_x_opt[x], plot_y_opt[x], label = label2)
    ax.legend(loc='upper left')
    ax.set_xlabel('Number of queries')
    ax.set_ylabel('Time in ms')
    
plt.savefig('graphs_best_vers.png')
plt.show()
```
